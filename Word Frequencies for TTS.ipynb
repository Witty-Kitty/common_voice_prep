```python
while("" in flat_list) :
    flat_list.remove("")
```

```python
# 5
lowered_list = [string.lower() for string in flat_list]
#6
clean_list = [''.join(letter for letter in word if letter not in string.punctuation) for word in lowered_list]
# 7
word_freq = Counter(clean_list).most_common()
```

```python
with open('tts_frequencies.txt', 'w') as f:
    for item in word_freq:
        f.write(str(item) + '\n')
```
