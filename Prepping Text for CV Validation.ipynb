```python
import csv
```

```python
f = open("1.txt")

text = f.read()

f.close()

# print(text)
```

```python
# split at newline characters
text_list = text.splitlines()
```

```python
# remove empty strings from list
while("" in text_list) :
    text_list.remove("")
```

```python
# split at fullstops
nested_list = [string.split(".") for string in text_list]
```

```python
# flatten the now nested list
flat_list = [nest for sublist in nested_list for nest in sublist]
```

```python
# remove empty strings from list
while("" in flat_list) :
    flat_list.remove("")
```

```python
# make everything a flat nested list, for purposes of saving to csv
list_of_list = [[string] for string in flat_list]
```

```python
with open('1.csv', 'w', newline='') as csvfile:
    writer= csv.writer(csvfile, delimiter=' ')
    writer.writerows(list_of_list)
```
